```python
df = pd.DataFrame({'key1': ['a', 'a', 'a', 'b', 'b'],
                  'key2': ['one', 'one', 'two', 'one', 'two'],
                  'data1': np.random.randn(5),
                  'data2': np.random.randn(5)})
df
```

```python
df.groupby(['key1','key2']).min()
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()
```

```python
a = np.array([[1,2],[3,4]])
a
```

```python
peak_to_peak(a)
```

```python
df.groupby('key1').agg(peak_to_peak)
```
