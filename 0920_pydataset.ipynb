# sns.countplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
sns.countplot(data = df, x = 'sex')
```

```python
sns.countplot(data = df, x = 'sex', hue = 'alive')
```

```python
sns.countplot(data = df, y = 'sex', hue = 'alive')
```
