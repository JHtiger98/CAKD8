## IQR, 극단치 경계
- IQR: 1사분위 ~ 3사분위
- 상한(윗수염의 끝): 3사분위 + 1.5IQR
- 하한(아랫수염의 끝): 1사분위 - 1.5IQR

```python
exam['math'].quantile(.75)
```

```python
exam['math'].quantile(.25)
```

```python
#IQR 구하기
pct75 = exam['math'].quantile(.75)
pct25 = exam['math'].quantile(.25)
IQR = pct75-pct25
IQR
```

```python
#상한 구하기
pct75 + 1.5*IQR
```

```python
#하한 구하기
pct25 - 1.5*IQR
```

# DataFrame의 연산

```python
import numpy as np
```

```python
frame = pd.DataFrame(np.random.randn(4, 3), index=['Utah', 'Ohio', 'Texas', 'Oregon'], columns=['b','d','e'])
frame
```

```python
frame = frame.abs()
frame
```

```python
#행(x축)을 기준으로 더해준다
frame.sum(axis = 0)
```

```python
#열(y축)을 기준으로 더해준다
frame.sum(axis = 1)
```

```python
#행(x축)을 기준으로 평균을 구해준다
frame.mean(axis = 0)
```

```python
#열(y축)을 기준으로 평균을 구해준다
frame.mean(axis = 1)
```

```python
frame.apply(lambda x: x.max() - x.min(), axis=0)
```

```python
frame.apply(lambda x: x.max() - x.min(), axis=1)
```

```python
#반올림
form = lambda x: '%.2f' %x
frame.applymap(form)
```

```python
#행(x축)을 기준으로 랭킹 매기기
frame.rank(axis = 0)
```

```python
#열(y축) 기준으로 랭킹 매기기
frame.rank(axis = 1)
```
